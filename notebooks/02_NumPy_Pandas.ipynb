## PARTIE 1 — NumPy

### 1.1 Création d'arrays

```python
# Depuis une liste Python
arr1d = np.array([1, 2, 3, 4, 5])
arr2d = np.array([[1, 2, 3], [4, 5, 6]])

print(arr1d)
print(arr2d)
print("Shape:", arr2d.shape)    # (lignes, colonnes)
print("Dtype:", arr2d.dtype)
print("Ndim:", arr2d.ndim)
print("Size:", arr2d.size)      # nombre total d'éléments

# Fonctions utiles de création
print(np.zeros((2, 3)))         # rempli de 0
print(np.ones((2, 3)))          # rempli de 1
print(np.eye(3))                # matrice identité
print(np.arange(0, 10, 2))      # comme range()
print(np.linspace(0, 1, 5))     # 5 valeurs entre 0 et 1
```

### 1.2 Indexing & Slicing

```python
arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print(arr[0, 2])         # ligne 0, colonne 2 → 30
print(arr[1, :])         # toute la ligne 1 → [40 50 60]
print(arr[:, 0])         # toute la colonne 0 → [10 40 70]
print(arr[0:2, 1:3])     # sous-matrice

# Boolean indexing
masque = arr > 50
print(masque)
print(arr[masque])       # valeurs > 50

# Fancy indexing
print(arr[[0, 2], :])    # lignes 0 et 2
```

### 1.3 Opérations et broadcasting

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

print(a + b)          # opérations élément par élément
print(a * 2)          # broadcasting : multiplie chaque élément
print(a ** 2)

# Agrégations
print(np.sum(a))
print(np.mean(a))
print(np.std(a))
print(np.max(a), np.min(a))
print(np.cumsum(a))   # somme cumulée

# Sur des matrices : axis=0 (par colonne), axis=1 (par ligne)
m = np.array([[1, 2, 3], [4, 5, 6]])
print(np.sum(m, axis=0))   # somme de chaque colonne
print(np.sum(m, axis=1))   # somme de chaque ligne
```

### 1.4 Reshape, stack, split

```python
arr = np.arange(12)
print(arr.reshape(3, 4))    # 3 lignes, 4 colonnes
print(arr.reshape(2, 2, 3)) # tensor 3D
print(arr.reshape(4, -1))   # -1 = calculé automatiquement

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.vstack([a, b]))   # empilement vertical
print(np.hstack([a, b]))   # empilement horizontal
print(np.concatenate([a, b]))
```

### 1.5 Nombres aléatoires

```python
rng = np.random.default_rng(42)  # seed pour reproducibilité

print(rng.integers(0, 10, 5))          # entiers entre 0 et 10
print(rng.random(5))                   # flottants entre 0 et 1
print(rng.normal(loc=0, scale=1, size=5))  # distribution normale
print(rng.choice(["a", "b", "c"], size=10, replace=True))  # tirage
```

## PARTIE 2 — Pandas

### 2.1 Series et DataFrame

```python
# Series : colonne 1D avec index
s = pd.Series([10, 20, 30, 40], index=["a", "b", "c", "d"])
print(s)
print(s["b"])     # accès par label
print(s[1])       # accès par position
print(s > 15)     # filtrage booléen

# DataFrame : table 2D
data = {
    "nom": ["Alice", "Bob", "Charlie", "Diana"],
    "age": [28, 34, 22, 45],
    "score": [87.5, 92.0, 78.3, 95.1],
    "ville": ["Paris", "Lyon", "Paris", "Marseille"]
}
df = pd.DataFrame(data)
print(df)
```

### 2.2 Exploration d'un DataFrame

```python
print(df.shape)          # (lignes, colonnes)
print(df.dtypes)         # types de colonnes
print(df.info())         # résumé complet
print(df.describe())     # statistiques descriptives
print(df.head(2))        # 2 premières lignes
print(df.tail(2))        # 2 dernières lignes
print(df.columns.tolist())
print(df.index)
```

### 2.3 Sélection de données

```python
# Sélection de colonnes
print(df["nom"])                     # une colonne (Series)
print(df[["nom", "score"]])           # plusieurs colonnes (DataFrame)

# .loc : sélection par label
print(df.loc[0, "nom"])              # cellule précise
print(df.loc[0:2, "nom":"score"])    # plage de lignes et colonnes

# .iloc : sélection par position entière
print(df.iloc[0, 0])                 # ligne 0, colonne 0
print(df.iloc[1:3, :2])              # lignes 1-2, colonnes 0-1

# Filtrage conditionnel
print(df[df["age"] > 30])
print(df[(df["ville"] == "Paris") & (df["score"] > 85)])

# .query() : plus lisible
print(df.query("age > 25 and score > 80"))
```

### 2.4 Manipulation de données

```python
# Ajouter une colonne
df["score_normalise"] = (df["score"] - df["score"].mean()) / df["score"].std()

# Modifier des valeurs avec .loc
df.loc[df["nom"] == "Bob", "ville"] = "Bordeaux"

# Renommer des colonnes
df_renomme = df.rename(columns={"nom": "prenom", "score": "note"})
print(df_renomme.columns.tolist())

# Supprimer des colonnes
df_reduit = df.drop(columns=["score_normalise"])

# Trier
print(df.sort_values("score", ascending=False))
print(df.sort_values(["ville", "age"]))  # tri multi-colonnes

# Appliquer une fonction
df["mention"] = df["score"].apply(lambda s: "Bien" if s >= 85 else "Passable")
print(df)
```

### 2.5 GroupBy et agrégations

```python
# GroupBy basique
print(df.groupby("ville")["score"].mean())

# Plusieurs agrégations à la fois
print(df.groupby("ville").agg(
    nb_personnes=("nom", "count"),
    score_moyen=("score", "mean"),
    age_max=("age", "max")
))

# transform() : conserve l'index original
df["score_moyen_ville"] = df.groupby("ville")["score"].transform("mean")
print(df[["nom", "ville", "score", "score_moyen_ville"]])
```

### 2.6 Fusion de DataFrames (Merge/Join)

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "nom": ["Alice", "Bob", "Charlie", "Diana"]
})
df2 = pd.DataFrame({
    "id": [1, 2, 3, 5],
    "score": [87, 92, 78, 95]
})

# INNER JOIN : seulement les correspondances
inner = pd.merge(df1, df2, on="id", how="inner")
print("INNER:\n", inner)

# LEFT JOIN : tous les de gauche, NaN si pas de correspondance
left = pd.merge(df1, df2, on="id", how="left")
print("LEFT:\n", left)

# OUTER JOIN : tous les enregistrements
outer = pd.merge(df1, df2, on="id", how="outer")
print("OUTER:\n", outer)

# concat() : empiler plusieurs DataFrames
df_combined = pd.concat([df1, df1], ignore_index=True)
print(df_combined)
```

### 2.7 Données manquantes dans Pandas

```python
df_nan = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [np.nan, 2, 3, 4],
    "C": [1, 2, 3, 4]
})

print(df_nan.isnull())           # masque booléen
print(df_nan.isnull().sum())     # nb NaN par colonne
print(df_nan.isnull().mean())    # proportion de NaN par colonne

# Supprimer
print(df_nan.dropna())           # supprime lignes avec au moins un NaN
print(df_nan.dropna(axis=1))     # supprime colonnes

# Remplir
print(df_nan.fillna(0))                            # remplace par 0
print(df_nan.fillna(df_nan.mean()))                # remplace par la moyenne
print(df_nan.fillna(method="ffill"))               # forward fill
```

### 2.8 Pivot Tables & Reshaping

```python
# Données : ventes par vendeur et produit
ventes = pd.DataFrame({
    "vendeur": ["Alice", "Alice", "Bob", "Bob", "Alice"],
    "produit": ["A", "B", "A", "B", "A"],
    "montant": [100, 200, 150, 250, 120]
})

# Pivot table
pivot = ventes.pivot_table(
    values="montant",
    index="vendeur",
    columns="produit",
    aggfunc="sum",
    fill_value=0
)
print(pivot)

# melt() : wide → long format
pivot_reset = pivot.reset_index()
long = pivot_reset.melt(id_vars="vendeur", var_name="produit", value_name="total")
print(long)
```

## Exercices de révision

```python
# Données pour les exercices
import numpy as np
import pandas as pd

np.random.seed(0)
df_ex = pd.DataFrame({
    "departement": np.random.choice(["IT", "Marketing", "RH", "Finance"], 50),
    "salaire": np.random.normal(45000, 10000, 50).round(0),
    "anciennete": np.random.randint(0, 20, 50),
    "satisfaction": np.random.choice([1, 2, 3, 4, 5], 50),
})
# Introduire des NaN
df_ex.loc[np.random.choice(df_ex.index, 5), "salaire"] = np.nan
print(df_ex.head())
```

```python
# EXERCICE 1 : Quel est le salaire moyen par département ?
# (ignorer les NaN, utiliser groupby)

# À compléter !

# EXERCICE 2 : Créer une nouvelle colonne 'categorie_salaire'
# 'bas' si < 40000, 'moyen' si 40000-55000, 'haut' si > 55000

# À compléter !

# EXERCICE 3 : Créer une array NumPy 5x5 de 0 à 24
# et calculer la somme de chaque colonne

# À compléter !
```

```python
# --- SOLUTIONS ---

# Ex 1
print("Salaire moyen par département:")
print(df_ex.groupby("departement")["salaire"].mean().round(0))

# Ex 2
def categoriser_salaire(s):
    if pd.isna(s): return "inconnu"
    if s < 40000: return "bas"
    if s <= 55000: return "moyen"
    return "haut"

df_ex["categorie_salaire"] = df_ex["salaire"].apply(categoriser_salaire)
print(df_ex["categorie_salaire"].value_counts())

# Ex 3
m = np.arange(25).reshape(5, 5)
print(m)
print("Somme colonnes:", np.sum(m, axis=0))
```
