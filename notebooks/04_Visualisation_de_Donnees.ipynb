# 📊 Visualisation de Données
**Cours couverts :** Understanding Data Visualization, Introduction to Data Visualization with Matplotlib, Introduction to Data Visualization with Seaborn, Introduction to Data Visualization with Plotly in Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données de démonstration
np.random.seed(42)
n = 200

df = pd.DataFrame({
    "age": np.random.normal(35, 10, n).astype(int),
    "salaire": np.random.normal(50000, 15000, n),
    "experience": np.random.randint(0, 30, n),
    "departement": np.random.choice(["IT", "Marketing", "RH", "Finance", "Ventes"], n),
    "satisfaction": np.random.randint(1, 6, n),
    "genre": np.random.choice(["F", "M"], n),
})
df["salaire"] = df["salaire"] + df["experience"] * 1000
df["salaire"] = df["salaire"].clip(lower=20000)

# Données temporelles
dates = pd.date_range("2022-01", periods=24, freq="ME")
df_temps = pd.DataFrame({
    "date": dates,
    "ventes": 1000 + np.cumsum(np.random.randn(24) * 50) + np.linspace(0, 300, 24),
    "couts": 800 + np.cumsum(np.random.randn(24) * 30) + np.linspace(0, 100, 24)
})

print("Données créées. Shape:", df.shape)
```

---
## PARTIE 1 — Matplotlib

### 1.1 Anatomie d'un graphique Matplotlib

```python
# Figure / Axes / Artist
# - Figure : le conteneur principal (la feuille)
# - Axes   : un sous-graphique sur la figure
# - Artist : tout ce qu'on voit (lignes, textes, etc.)

fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x), label="sin(x)", color="steelblue", linewidth=2)
ax.plot(x, np.cos(x), label="cos(x)", color="coral",    linewidth=2, linestyle="--")

ax.set_title("Fonctions trigonométriques", fontsize=14, fontweight="bold")
ax.set_xlabel("x (radians)")
ax.set_ylabel("Valeur")
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color="black", linewidth=0.5)  # ligne horizontale à y=0

plt.tight_layout()
plt.show()
```

### 1.2 Subplots (plusieurs graphiques)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Histogramme
axes[0, 0].hist(df["salaire"], bins=30, color="steelblue", edgecolor="white", alpha=0.8)
axes[0, 0].set_title("Distribution des salaires")
axes[0, 0].set_xlabel("Salaire (€)")
axes[0, 0].axvline(df["salaire"].mean(), color="red", linestyle="--", label=f"Moyenne: {df['salaire'].mean():.0f}")
axes[0, 0].legend()

# Scatter plot
axes[0, 1].scatter(df["experience"], df["salaire"], alpha=0.4, c="coral")
axes[0, 1].set_title("Expérience vs Salaire")
axes[0, 1].set_xlabel("Années d'expérience")
axes[0, 1].set_ylabel("Salaire (€)")

# Bar chart
dept_avg = df.groupby("departement")["salaire"].mean().sort_values(ascending=False)
axes[1, 0].bar(dept_avg.index, dept_avg.values, color="mediumseagreen", edgecolor="white")
axes[1, 0].set_title("Salaire moyen par département")
axes[1, 0].tick_params(axis="x", rotation=45)

# Line plot (série temporelle)
axes[1, 1].plot(df_temps["date"], df_temps["ventes"], label="Ventes", color="steelblue")
axes[1, 1].plot(df_temps["date"], df_temps["couts"], label="Coûts", color="coral")
axes[1, 1].fill_between(df_temps["date"], df_temps["ventes"], df_temps["couts"], alpha=0.2, color="green")
axes[1, 1].set_title("Ventes et coûts dans le temps")
axes[1, 1].legend()
axes[1, 1].tick_params(axis="x", rotation=45)

fig.suptitle("Tableau de bord RH", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

### 1.3 Types de graphiques avancés

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Pie chart
dept_count = df["departement"].value_counts()
axes[0].pie(dept_count.values, labels=dept_count.index,
            autopct="%1.1f%%", startangle=90)
axes[0].set_title("Répartition par département")

# Boxplot
depts = df["departement"].unique()
data_box = [df[df["departement"] == d]["salaire"].values for d in depts]
axes[1].boxplot(data_box, labels=depts)
axes[1].set_title("Distribution salaires par dépt")
axes[1].tick_params(axis="x", rotation=45)
# Boxplot : médiane (ligne orange), Q1-Q3 (boîte), moustaches (1.5*IQR), points = outliers

# Heatmap avec imshow
corr_data = df[["age", "salaire", "experience", "satisfaction"]].corr()
im = axes[2].imshow(corr_data, cmap="coolwarm", vmin=-1, vmax=1)
axes[2].set_xticks(range(len(corr_data.columns)))
axes[2].set_yticks(range(len(corr_data.columns)))
axes[2].set_xticklabels(corr_data.columns, rotation=45)
axes[2].set_yticklabels(corr_data.columns)
plt.colorbar(im, ax=axes[2])
axes[2].set_title("Matrice de corrélation")

plt.tight_layout()
plt.show()
```

## PARTIE 2 — Seaborn

Seaborn = Matplotlib avec une syntaxe plus simple et de plus beaux styles par défaut.

```python
sns.set_theme(style="whitegrid", palette="husl")

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# histplot
sns.histplot(data=df, x="salaire", kde=True, ax=axes[0, 0])
axes[0, 0].set_title("histplot : distribution + KDE")

# boxplot
sns.boxplot(data=df, x="departement", y="salaire", ax=axes[0, 1])
axes[0, 1].set_title("boxplot")
axes[0, 1].tick_params(axis="x", rotation=45)

# violinplot (boxplot + distribution)
sns.violinplot(data=df, x="departement", y="salaire", ax=axes[0, 2])
axes[0, 2].set_title("violinplot")
axes[0, 2].tick_params(axis="x", rotation=45)

# scatterplot avec hue
sns.scatterplot(data=df, x="experience", y="salaire", hue="genre",
                alpha=0.6, ax=axes[1, 0])
axes[1, 0].set_title("scatterplot avec hue")

# barplot (moyenne + IC)
sns.barplot(data=df, x="departement", y="salaire", estimator="mean",
            errorbar="ci", ax=axes[1, 1])
axes[1, 1].set_title("barplot (moyenne + IC 95%)")
axes[1, 1].tick_params(axis="x", rotation=45)

# heatmap
corr = df[["age", "salaire", "experience", "satisfaction"]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
            center=0, ax=axes[1, 2])
axes[1, 2].set_title("heatmap corrélation")

plt.suptitle("Galerie Seaborn", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# FacetGrid : grilles de graphiques selon une variable
g = sns.FacetGrid(df, col="departement", col_wrap=3, height=3)
g.map(sns.histplot, "salaire", bins=15, kde=True)
g.set_titles("{col_name}")
g.set_axis_labels("Salaire", "Effectif")
plt.suptitle("Distribution des salaires par département", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# pairplot : relations entre toutes les variables numériques
g = sns.pairplot(df[["age", "salaire", "experience", "satisfaction"]].sample(80),
                 diag_kind="kde", plot_kws={"alpha": 0.4})
plt.suptitle("Pairplot", y=1.01)
plt.show()
```

## PARTIE 3 — Plotly (visualisations interactives)

```python
try:
    import plotly.express as px
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    PLOTLY = True
except ImportError:
    print("Plotly non installé. Exécutez : pip install plotly")
    PLOTLY = False
```

```python
if PLOTLY:
    # Scatter interactif
    fig = px.scatter(
        df, x="experience", y="salaire",
        color="departement", size="age",
        hover_data=["genre", "satisfaction"],
        title="Expérience vs Salaire (interactif)",
        labels={"experience": "Années d'expérience", "salaire": "Salaire (€)"}
    )
    fig.show()
```

```python
if PLOTLY:
    # Box plot par département
    fig = px.box(df, x="departement", y="salaire", color="genre",
                 title="Salaires par département et genre")
    fig.show()

    # Ligne temporelle
    fig2 = px.line(df_temps, x="date", y=["ventes", "couts"],
                   title="Évolution des ventes et coûts")
    fig2.show()
```

```python
if PLOTLY:
    # Subplots Plotly
    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=("Histogramme", "Pie chart"))

    fig.add_trace(go.Histogram(x=df["salaire"], name="Salaire",
                                marker_color="steelblue"), row=1, col=1)

    dept_cnt = df["departement"].value_counts()
    fig.add_trace(go.Pie(labels=dept_cnt.index, values=dept_cnt.values,
                          name="Départements"), row=1, col=2)

    fig.update_layout(title="Dashboard interactif", showlegend=False)
    fig.show()
```

## PARTIE 4 — Principes de bonne visualisation

```python
principes = """
CHOISIR LE BON TYPE DE GRAPHIQUE :

  • Distribution       → Histogramme, KDE plot, Boxplot, Violin plot
  • Comparaison        → Bar chart, Grouped bar, Boxplot
  • Relation/Corrélation → Scatter plot, Heatmap, Bubble chart
  • Composition        → Pie chart, Stacked bar, Treemap
  • Évolution temporelle → Line chart, Area chart

BONNES PRATIQUES :
  1. Toujours labeliser les axes et le titre
  2. Commencer l'axe Y à 0 pour les bar charts (sauf exceptions)
  3. Éviter le pie chart avec > 5 catégories
  4. Utiliser des couleurs accessibles (daltonisme : éviter rouge/vert seuls)
  5. Limiter le nombre de couleurs (max 5-7 catégories)
  6. Éviter le 3D qui déforme la perception
  7. Data-ink ratio : enlever le superflu (grid, bordures, etc.)
  8. Annoter les points importants
"""
print(principes)
```

```python
# --- SOLUTIONS ---

# Ex 1
sal_exp = df.groupby("tranche_exp", observed=True)["salaire"].agg(["mean", "std"]).reset_index()

fig, ax = plt.subplots(figsize=(9, 4))
ax.bar(sal_exp["tranche_exp"].astype(str), sal_exp["mean"],
       yerr=sal_exp["std"], capsize=5, color="steelblue", edgecolor="white")
ax.set_title("Salaire moyen par tranche d'expérience (± 1 écart-type)")
ax.set_xlabel("Expérience (années)")
ax.set_ylabel("Salaire moyen (€)")
plt.tight_layout()
plt.show()

# Ex 2
pivot = df.pivot_table(values="satisfaction", index="departement",
                       columns="tranche_exp", aggfunc="mean", observed=True)
fig, ax = plt.subplots(figsize=(10, 4))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="RdYlGn",
            vmin=1, vmax=5, ax=ax)
ax.set_title("Satisfaction moyenne par département et expérience")
plt.tight_layout()
plt.show()
```
