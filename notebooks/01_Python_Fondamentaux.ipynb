## 1. Types de données de base

```python
# --- Types de base ---
entier = 42
flottant = 3.14
texte = "Bonjour"
booleen = True

print(type(entier), type(flottant), type(texte), type(booleen))

# Conversion de types (casting)
print(int(3.9))      # → 3 (tronque, ne pas arrondir)
print(float("7"))   # → 7.0
print(str(100))      # → '100'
```

## 2. Listes, Tuples, Dictionnaires, Ensembles

```python
# --- LISTES (mutables, ordonnées) ---
fruits = ["pomme", "banane", "cerise"]
fruits.append("mangue")          # ajout en fin
fruits.insert(1, "fraise")       # ajout à l'index 1
fruits.remove("banane")          # suppression par valeur
print(fruits)

# Slicing
print(fruits[1:3])   # index 1 inclus, 3 exclus
print(fruits[::-1])  # liste inversée

# List comprehension
carres = [x**2 for x in range(1, 6)]
print(carres)
```

```python
# --- TUPLES (immutables, ordonnés) ---
coordonnees = (48.8566, 2.3522)   # Paris
lat, lon = coordonnees            # dépackaging
print(f"Latitude: {lat}, Longitude: {lon}")

# --- DICTIONNAIRES (clé-valeur, non ordonnés avant Python 3.7) ---
personne = {"nom": "Alice", "age": 28, "ville": "Paris"}
personne["email"] = "alice@example.com"   # ajout
print(personne.get("age"))                # accès sûr
print(list(personne.keys()))
print(list(personne.values()))

# Dict comprehension
carres_dict = {x: x**2 for x in range(1, 6)}
print(carres_dict)
```

```python
# --- ENSEMBLES (sets) : pas de doublons, non ordonnés ---
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union
print(a & b)   # intersection
print(a - b)   # différence (dans a mais pas dans b)
print(a ^ b)   # différence symétrique
```

## 3. Conditions et boucles

```python
# --- Conditions ---
note = 75

if note >= 90:
    mention = "Très bien"
elif note >= 70:
    mention = "Bien"
elif note >= 50:
    mention = "Passable"
else:
    mention = "Insuffisant"

print(f"Note: {note} → Mention: {mention}")

# Opérateur ternaire
statut = "majeur" if note >= 50 else "mineur"
print(statut)
```

```python
# --- Boucles ---

# for avec range
for i in range(5):
    print(i, end=" ")
print()

# for avec enumerate (index + valeur)
langages = ["Python", "SQL", "R"]
for idx, lang in enumerate(langages):
    print(f"{idx}: {lang}")

# for avec zip (deux listes en parallèle)
noms = ["Alice", "Bob", "Charlie"]
ages = [28, 34, 22]
for nom, age in zip(noms, ages):
    print(f"{nom} a {age} ans")
```

```python
# --- Boucle while ---
compteur = 0
while compteur < 5:
    print(compteur, end=" ")
    compteur += 1
print()

# break et continue
for i in range(10):
    if i == 3:
        continue   # passe à l'itération suivante
    if i == 7:
        break      # arrête la boucle
    print(i, end=" ")
```

## 4. Fonctions

```python
# --- Fonction de base ---
def calculer_imc(poids, taille):
    """Calcule l'IMC d'une personne.
    
    Args:
        poids: poids en kg
        taille: taille en mètres
    Returns:
        IMC arrondi à 2 décimales
    """
    imc = poids / taille**2
    return round(imc, 2)

print(calculer_imc(70, 1.75))    # arguments positionnels
print(calculer_imc(taille=1.68, poids=60))  # arguments nommés
```

```python
# --- Arguments par défaut ---
def saluer(nom, langue="fr"):
    messages = {"fr": "Bonjour", "en": "Hello", "es": "Hola"}
    return f"{messages.get(langue, 'Hi')}, {nom}!"

print(saluer("Alice"))
print(saluer("Bob", "en"))

# --- *args : nombre variable d'arguments positionnels ---
def somme(*nombres):
    return sum(nombres)

print(somme(1, 2, 3, 4, 5))

# --- **kwargs : nombre variable d'arguments nommés ---
def afficher_infos(**infos):
    for cle, valeur in infos.items():
        print(f"  {cle}: {valeur}")

afficher_infos(nom="Alice", age=28, ville="Paris")
```

```python
# --- Fonctions lambda (anonymes) ---
doubler = lambda x: x * 2
print(doubler(5))

# Très utiles avec map(), filter(), sorted()
nombres = [3, 1, 4, 1, 5, 9, 2, 6]
print(sorted(nombres, reverse=True))

pairs = list(filter(lambda x: x % 2 == 0, nombres))
print(pairs)

doubles = list(map(lambda x: x * 2, nombres))
print(doubles)
```

```python
# --- Scope : variables locales vs globales ---
x = 10   # variable globale

def modifier():
    x = 99   # variable LOCALE, ne modifie pas le x global
    print("Dedans:", x)

modifier()
print("Dehors:", x)  # toujours 10

# Pour modifier le global, utiliser global (à éviter en pratique)
def modifier_global():
    global x
    x = 99

modifier_global()
print("Après global:", x)
```

## 5. Programmation Orientée Objet (OOP)

```python
# --- Classe de base ---
class Animal:
    # Attribut de classe (partagé par toutes les instances)
    espece = "Animalia"
    
    def __init__(self, nom, age):
        # Attributs d'instance
        self.nom = nom
        self.age = age
    
    def parler(self):
        return f"{self.nom} fait un son."
    
    def __str__(self):
        return f"{self.nom} ({self.age} ans)"
    
    def __repr__(self):
        return f"Animal(nom={self.nom!r}, age={self.age!r})"


# --- Héritage ---
class Chien(Animal):
    def __init__(self, nom, age, race):
        super().__init__(nom, age)   # appel du parent
        self.race = race
    
    def parler(self):   # surcharge de méthode
        return f"{self.nom} aboie : Ouaf!"


rex = Chien("Rex", 3, "Berger Allemand")
print(rex)
print(rex.parler())
print(isinstance(rex, Animal))   # True (héritage)
```

## 6. Gestion des erreurs

```python
# --- Try / Except ---
def diviser(a, b):
    try:
        resultat = a / b
    except ZeroDivisionError:
        return "Erreur : division par zéro!"
    except TypeError as e:
        return f"Erreur de type: {e}"
    else:
        return resultat    # exécuté si pas d'erreur
    finally:
        print("Opération terminée.")  # TOUJOURS exécuté

print(diviser(10, 2))
print(diviser(10, 0))
print(diviser(10, "a"))
```

## 7. Itérateurs et générateurs

```python
# --- Générateurs : économisent de la mémoire ---
def compter_jusqu(n):
    """Générateur qui produit des nombres un à un."""
    for i in range(n):
        yield i  # 'yield' suspend la fonction et retourne la valeur

gen = compter_jusqu(5)
print(next(gen))  # 0
print(next(gen))  # 1

# Itérer tout le générateur
for val in compter_jusqu(5):
    print(val, end=" ")
print()

# Generator expression (comme list comp mais paresseux)
gen_carres = (x**2 for x in range(10))
print(sum(gen_carres))  # ne charge pas tout en mémoire
```

```python
# --- SOLUTIONS ---

# Solution Exercice 1
import statistics

def statistiques_sol(nombres):
    return {
        'min': min(nombres),
        'max': max(nombres),
        'moyenne': round(sum(nombres) / len(nombres), 2),
        'mediane': statistics.median(nombres)
    }

print(statistiques_sol([4, 7, 2, 9, 1, 5, 8]))

# Solution Exercice 2
class CompteBancaireSol:
    def __init__(self, proprietaire, solde_initial=0):
        self.proprietaire = proprietaire
        self._solde = solde_initial
    
    def deposer(self, montant):
        if montant <= 0:
            raise ValueError("Le montant doit être positif.")
        self._solde += montant
    
    def retirer(self, montant):
        if montant > self._solde:
            raise ValueError("Solde insuffisant.")
        self._solde -= montant
    
    def solde(self):
        return self._solde

compte = CompteBancaireSol("Alice", 1000)
compte.deposer(500)
print(compte.solde())

# Solution Exercice 3
fizzbuzz_sol = [x for x in range(1, 101) if (x % 3 == 0) ^ (x % 5 == 0)]
print(fizzbuzz_sol[:10])
```
